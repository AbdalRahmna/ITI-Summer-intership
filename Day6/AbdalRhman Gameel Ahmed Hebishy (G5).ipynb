## Define & Remove the outliers :

```python
# Select numerical features
features = df.select_dtypes(include="number").columns


columns_with_outliers = []


for col in features:
    Q1_col, Q3_col = df[col].quantile([0.25, 0.75])
    iqr = Q3_col - Q1_col
    low_limit = Q1_col - 1.5 * iqr
    upper_limit = Q3_col + 1.5 * iqr
    
    
    df_no_outliers = df[(df[col] >= low_limit) & (df[col] <= upper_limit)]
    
 
    if len(df_no_outliers) < len(df):
        columns_with_outliers.append(col)
        print(f"There are outliers in the feature {col}")
    else:
        print(f"There are no outliers in the feature {col}")
    
    
    df = df_no_outliers
    
   
    print(f"Q1 of {col} --->>> {Q1_col}")
    print(f"Q3 of {col} --->>> {Q3_col}")
    print(f"IQR --->>> {iqr}")
    print(f"Low limit --->>> {low_limit}")
    print(f"Upper limit --->>> {upper_limit}")
    print(f"Number of outliers removed --->>> {len(df) - len(df_no_outliers)}")
    print("--------------------------------------------------")

# Display the first few rows of the cleaned data
df.head()
```

```python
f1={"family":"serif","size":9,"color":"r"}
plt.figure(figsize=(12,17),dpi=250)
plt.title("This Represention to check if found  outliers",fontdict=f1)
plt.xlabel("Features",fontdict=f1)
plt.ylabel("count of happens",fontdict=f1)
plt.xticks(rotation=45,color="b")
sns.boxplot(data[features])
plt.legend(title="outliers",prop={'size': 5})
```
